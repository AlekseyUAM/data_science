```python
import csv, random, pandas, seaborn
```

```python
names = []
with open('t1', 'r') as f:
    for line in f:
        names.append(line[:28].strip())
```

```python
header = ['Магазин']
for i in range(14):
    header.append('Неделя ' + str(i+1))
data = [[random.randint(0, 100)/100 for j in range(14)] for i in range(38)]
for i,v in enumerate(data):
    v.insert(0, names[i])
```

```python
with open("polomki.csv", "wt") as fp:
    writer = csv.writer(fp, delimiter=",")
    writer.writerow(header)  # write header
    writer.writerows(data)
```

```python
data = pandas.read_csv('polomki.csv', index_col='Магазин')
data['Неделя 14'] *= 10 # умножение значений столбца
```

```python
seaborn.heatmap(data)
```
